```python
import requests as re
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 100)
```

```python
#acled API request
data = re.get('https://api.acleddata.com/acled/read?terms=accept&country=%Nigeria%&limit=0&event_date=2009-01-01|2019-01-30&event_date_where=BETWEEN')
```

```python
records = data.json()['data']
```

```python
len(records)
```

```python
#search records for mention of schools
schools = []
for record in records:
    if "school" in record['notes']:
        schools.append(record)
    
```

```python
len(schools)
```

```python
df = pd.DataFrame(schools)
```

```python
#filter for Boko Haram incidents
bh_school_attacks = df[(df['actor1'].str.contains('Boko Haram')) | (df['actor2'].str.contains('Boko Haram')) | (df['assoc_actor_2'].str.contains('Boko Haram'))]
```
